# DESCRIPTIVE ANALYSIS OF THE DATASET

```python
# WHAT IS THE DISTRIBUTION OF TV ADVERTISING SPENDING?

plt.figure(figsize=(10,5))
sns.histplot(data['TV'], bins=5)
plt.title('TV Advertising  Spending Distribution')
plt.xlabel('TV Advertising Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
# TV values ranging from 175 to 240 have the highest frequency of 48 
# TV values ranging from 120 to 180 have the lowest frequency of 35
```

```python
# WHAT IS THE DISTRIBUTION OF RADIO ADVERTISING SPENDING?

plt.figure(figsize=(10,5))
sns.histplot(data['Radio'], bins=5)
plt.title('Radio Advertising Spending Distribution')
plt.xlabel('Radio Advertising Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
# Radio values ranging from 0 to 10 have the highest frequency of 50 
# Radio values ranging from 30 to 39 have the lowest frequency of 34
```

```python
# WHAT IS THE DISTRIBUTION OF NEWSPAPER ADVERTISING SPENDING?

plt.figure(figsize=(10,5))
sns.histplot(data['Newspaper'], bins=5)
plt.title('Newspaper Advertising Spending Distribution')
plt.xlabel('Newspaper Advertising Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
# Newspaper values ranging from o to 19 have the highest frequency of 84 
# Newspaper values ranging from 86 to 170 have the lowest frequency of 0.1
```

```python
# WHAT IS THE DISTRIBUTION OF SALES FIGURE?

plt.figure(figsize=(10,5))
sns.histplot(data['Sales'], bins=5)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sales values ranging from 17 to 22 have the highest frequency of 64.
# Sales values ranging from 1.2 to 7 have the lowest frequency of 11.
```

```python
# To visualize distributions using histogram
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.hist(data['TV'], bins=10, color='red', alpha=1)
plt.title('TV Spending Distribution')
plt.xlabel('TV Spending')
plt.subplot(122)
plt.hist(data['Radio'], bins=10, color='purple', alpha=0.7)
plt.title('Radio Spending Distribution')
plt.xlabel('Radio Spending')
```

```python
# The above chart shows that TV spending values between 0 to 30 and 210 to 240 had the highest frequency in the TV and Radio column respectively 
```

```python
# To visualize distributions using histogram
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.hist(data['Newspaper'], bins=10, color='blue', alpha=0.7)
plt.title('Newspaper Spending Distribution')
plt.xlabel('Newspaper Spending')
plt.subplot(122)
plt.hist(data['Sales'], bins=10, color='green', alpha=0.7)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# The above chart shows that Newspaper spending and Sales values between 0 to 10 and 17 to 19 had the highest frequencies for Newspaper spending and Sales repectively
```

```python
# IS THERE A CORRELATION BETWEEN TV, RADIO, NEWSPAPER SPENDING AND SALES?

sns.heatmap(data.corr(), annot=True, cmap='viridis');
```

```python
# The heatmap above shows that; 
# TV has 90% relationship with Sales = (Positive correlation)
# Radio has 35% relationship with Sales = (Positive correlation)
# Newspaper has 16% relationship with Sales = (Positive correlation)
# Sales has 100% relationship with Sales(itself) = (Positive correlation)
```
